## Presentaion of rainfall data: Moving Average

```python
#For importing required modules

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

```python
dataset = pd.read_csv('RainfallData.txt',sep="     ",engine='python')
dataset.columns = ['Year', 'Mean_Rainfall(mm)']
dataset['moving_avg_3'] = dataset['Mean_Rainfall(mm)'].rolling(window=3).mean()
dataset['moving_avg_5'] = dataset['Mean_Rainfall(mm)'].rolling(window=5).mean()
dataset.head()

m5 = dataset['moving_avg_5']
m3 = dataset['moving_avg_3']

x = dataset['Year']
y = dataset['Mean_Rainfall(mm)']
```

```python
# For plotting

plt.figure(figsize=(20,15))

plt.subplot(221)
plt.plot(x,y)
plt.xlabel("Years")
plt.ylabel("Mean Rainfall(mm)")

plt.subplot(222)
plt.plot(x,m3,"r")
plt.xlabel("Years")
plt.ylabel("Rainfall(mm)")
plt.legend(["3 yrs moving average"])

plt.subplot(223)
plt.plot(x,m5,"g")
plt.xlabel("Years")
plt.ylabel("Rainfall(mm)")
plt.legend(["5 yrs moving average"])

plt.subplot(224)
plt.plot(x,y, x,m3,"r--", x,m5,"g--")
plt.xlabel("Years")
plt.ylabel("Rainfall(mm)")
plt.legend(['Mean Rainfall','3yr moving avg','5yr moving avg'])

plt.show()
```

The red curve is computed taking 3 years moving average.
the green curve is computed taking 5 years moving average.

## Presentation of rainfall data: Point rainfall

```python
mean_0 = dataset['Mean_Rainfall(mm)'].mean()
mean_p20 = mean_0 + (.2*mean_0)
mean_n20 = mean_0 - (.2*mean_0)

dataset['Mean'] = mean_0
dataset['Mean + 20%'] = mean_p20
dataset['Mean - 20%'] = mean_n20
dataset.head()
```

```python
p,q,r = dataset['Mean'],dataset['Mean + 20%'],dataset['Mean - 20%']
fig = plt.figure(figsize=(10, 8))
ax = fig.add_axes([0,0,1,1])
ax.bar(x,y,color ='c', label = 'Mean Rainfall')

plt.plot(x,p,"r",label = 'Mean')
plt.plot(x,q,"b--",label = 'Mean + 20%')
plt.plot(x,r,"k--",label = 'Mean - 20%')

plt.xlabel("Year")
plt.ylabel("Mean Rainfall(mm)")
plt.legend()
plt.show()
```
